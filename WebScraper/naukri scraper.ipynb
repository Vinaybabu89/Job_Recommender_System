```python
import pandas as pd
from selenium import webdriver
from selenium import *
import time
from selenium.common.exceptions import *
import requests
import datetime
```

```python
path="C:/Program Files (x86)/chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
```

```python
lst=[]
```

```python
Roles=[]
Company=[]
Experience=[]
Salary=[]
Location=[]
Description=[]
Skills=[]
Posted_Date=[]
TYPE=[]
Present_Date=[]
```

```python
url="https://www.naukri.com/data-scientist-jobs-?k=data%20scientist"
```

```python
while True:
    #response=requests.get(url)
    driver.get(url)
    time.sleep(3)
    #lst=driver.find_elements_by_xpath("//article[@class='jobTuple bgWhite br4 mb-8']")
    lst=driver.find_elements_by_css_selector(".jobTuple.bgWhite.br4.mb-8")
    #article class="jobTuple bgWhite br4 mb-8"
    
    for data in lst:
        try:
            role=data.find_element_by_css_selector("a.title.fw500.ellipsis").text
            company=data.find_element_by_css_selector("a.subTitle.ellipsis.fleft").text
            exp=data.find_element_by_css_selector(".ellipsis.fleft.fs12.lh16").text
            salary=data.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.salary").text
            description=data.find_element_by_css_selector(".job-description.fs12.grey-text").text
            location=data.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.location").text
            skills=data.find_element_by_css_selector(".tags.has-description").text
            job=data.find_element_by_css_selector(".jobType.type.fleft.br2.mr-8").text
            date=data.find_element_by_css_selector(".type.br2.fleft.grey").text
        except (NoSuchElementException,AttributeError):
            pass
        try:
            url=driver.find_element_by_css_selector("a.fright.fs14.btn-secondary.br2").get_property('href')
        except AttributeError:
            break
        today=datetime.datetime.today().strftime('%Y-%m-%d')
        Roles.append(role)
        Company.append(company)
        Experience.append(exp)
        Salary.append(salary)
        Location.append(location)
        Description.append(description)
        Skills.append(skills)
        Posted_Date.append(date)
        Present_Date.append(today)
        TYPE.append(job)
        
```

```python
jobs1={'ROLES':Roles,'Companies':Company,'Experience':Experience,
       'Salary':Salary,'Location':Location,
       'Description':Description,'Skills':Skills,'Posted_Date':Posted_Date,
       'Present_Date':Present_Date,'TYPE':TYPE}
jobs1
```

```python
df1=pd.DataFrame(jobs1)
df1
```
